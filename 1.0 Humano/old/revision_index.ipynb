```python
import pandas as pd
import re
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
data_1=unpickle('dataset_humano_respuestas_parte1.pkl')
data_2=unpickle('dataset_humano_respuestas_parte2.pkl')
data_3=unpickle('dataset_humano_respuestas_parte3.pkl')
data_4=unpickle('dataset_humano_respuestas_parte4.pkl')
data_5=unpickle('dataset_humano_respuestas_parte5.pkl')
```

# Correcion de cabeceras

```python
def convertir_cabecera(x):
    string_lista=re.findall('"([^"]*)"',x)
    return string_lista[0]

nuevas_cabeceras_1=[convertir_cabecera(x) for x in data_1.columns.tolist()]
nuevas_cabeceras_2=[convertir_cabecera(x) for x in data_2.columns.tolist()]
nuevas_cabeceras_3=[convertir_cabecera(x) for x in data_3.columns.tolist()]
nuevas_cabeceras_4=[convertir_cabecera(x) for x in data_4.columns.tolist()]
nuevas_cabeceras_5=[convertir_cabecera(x) for x in data_5.columns.tolist()]
```

```python
data_1.columns=nuevas_cabeceras_1
data_2.columns=nuevas_cabeceras_2
data_3.columns=nuevas_cabeceras_3
data_4.columns=nuevas_cabeceras_4
data_5.columns=nuevas_cabeceras_5
```

## Guardando datasets actualizados

```python
data_1.to_pickle('dataset_humano_respuestas_parte1_new.pkl')
data_2.to_pickle('dataset_humano_respuestas_parte2_new.pkl')
data_3.to_pickle('dataset_humano_respuestas_parte3_new.pkl')
data_4.to_pickle('dataset_humano_respuestas_parte4_new.pkl')
data_5.to_pickle('dataset_humano_respuestas_parte5_new.pkl')
```
